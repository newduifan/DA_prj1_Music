```python
import pandas as pd
import os
from termcolor import colored
from IPython.display import display, HTML
import numpy as np

# 1. ÂÆö‰ΩçÂà∞ music_artist ÁõÆÂΩï (Locate the music_artist directory)
current_dir = os.path.dirname(os.path.abspath('__file__'))
project_root = os.path.abspath(os.path.join(current_dir, ".."))
artist_data_dir = os.path.join(project_root, 'data', 'music_artist')

# 2. Ëé∑ÂèñËØ•ÁõÆÂΩï‰∏ãÊâÄÊúâÁöÑ CSV Êñá‰ª∂ (Get all CSV files in the directory)
# ‰ΩøÁî® listdir Âπ∂ËøáÊª§ÂêéÁºÄÂêç
artist_files = [f for f in os.listdir(artist_data_dir) if f.endswith('.csv')]

# 3. ÊâπÈáèÂØºÂÖ•Âà∞Â≠óÂÖ∏ (Batch import into a dictionary)
artist_dfs = {}

print(colored(f"Starting import from: {artist_data_dir}", 'blue'))

for file_name in artist_files:
    file_path = os.path.join(artist_data_dir, file_name)
    # ‰ΩøÁî®Êñá‰ª∂Âêç‰Ωú‰∏∫ Key
    key = file_name.replace('.csv', '')

    try:
        # ÂØºÂÖ•Êï∞ÊçÆ
        # Ê≥®ÊÑèÔºöËøô‰∫õÊñá‰ª∂ÁöÑÁªìÊûÑÂèØËÉΩ‰∏çÂêåÔºåÂª∫ËÆÆÂÖà‰∏çÊåáÂÆöÁâπÂÆöÁöÑ dtypeÔºåÊàñËÄÖÁî® low_memory=False
        artist_dfs[key] = pd.read_csv(file_path, low_memory=False)
        print(colored(f"‚úÖ Imported: {key} | Shape: {artist_dfs[key].shape}", 'green'))
    except Exception as e:
        print(colored(f"‚ùå Error importing {file_name}: {e}", 'red'))

# 4. Âø´ÈÄüÊü•ÁúãÁªìÊûú (Quickly verify results)
print(f"\nTotal artist files loaded: {len(artist_dfs)}")
```

```python
# Âæ™ÁéØÊü•Áúã artist_dfs Â≠óÂÖ∏ÈáåÁöÑÊØè‰∏Ä‰∏™Ë°®
for name, df in artist_dfs.items():
    # 1. ÂàõÂª∫‰∏Ä‰∏™ÈÜíÁõÆÁöÑ HTML ËìùËâ≤Ê†áÈ¢ò
    display(HTML(f"""
        <div style="background-color: #e6f3ff; padding: 10px; border-radius: 5px; border-left: 5px solid #2196F3; margin-top: 20px;">
            <h3 style="color: #0d47a1; margin: 0;">üìä Dataset: {name}</h3>
            <p style="margin: 5px 0 0 0; color: #555;">Shape: {df.shape[0]:,} rows √ó {df.shape[1]} columns</p>
        </div>
    """))

    # 2. ÊâìÂç∞ Basic Info (Êï∞ÊçÆÁ±ªÂûã„ÄÅÈùûÁ©∫ÂÄºÁªüËÆ°)
    print(colored(f"\n>>> [{name}] Info:", 'blue', attrs=['bold']))
    df.info()

    # 3. ÁæéÂåñÂ±ïÁ§∫ Describe (Êï∞ÂÄºÂàóÁªüËÆ°ÊëòË¶Å)
    # ‰ΩøÁî® .T (ËΩ¨ÁΩÆ) ÊòØÂõ†‰∏∫Â¶ÇÊûúÂàóÊï∞ÂæàÂ§öÔºåÊ®™ÂêëÊéíÂàó‰ºöÂæàÈöæÁúãÔºåËΩ¨ÁΩÆÂêéÊõ¥Á¨¶ÂêàÈòÖËØª‰π†ÊÉØ
    print(colored(f"\n>>> [{name}] Statistical Summary (Transposed):", 'blue', attrs=['bold']))
    display(df.describe(include='all').T)

    print("-" * 80)
```

```python
# clean query
# ÊâãÂä®ÊèêÂèñÂá∫ÁâπÂÆöÁöÑ DataFrame Âπ∂Ëµã‰∫àÊñ∞ÂêçÂ≠ó
df_global = artist_dfs['Global Music Artists']
df_query = artist_dfs['query']
df_spotify = artist_dfs['top10k-spotify-artist-metadata']

df_query = df_query.drop(columns=['item'])

df_query.columns = ['art_name', 'gender', 'area', 'birth', 'death']
# Â∞ÜÊó∂Èó¥ÈáçÁΩÆ‰∏∫ÂΩìÊó•Èõ∂ÁÇπ (Normalize to midnight)
df_query['birth'] = pd.to_datetime(df_query['birth'], errors='coerce').dt.normalize()
df_query['death'] = pd.to_datetime(df_query['death'], errors='coerce').dt.normalize()

df_query['gender'] = df_query['gender'].astype('category')
print(df_query['gender'].dtype)
```

```python
# clean global music artists
df_global = df_global.drop(columns=['artist_img', 'artist_id'])
df_global.dropna(axis=1, how='all', inplace=True)

df_global.columns = ['art_name', 'art_genre', 'area']
```

```python
# clean top10k spotify artist metadata
df_spotify = df_spotify.drop(columns=['index', 'district_1', 'district_2', 'district_3'])
df_spotify.dropna(axis=1, how='all', inplace=True)

df_spotify.drop(columns=['Unnamed: 0', '<anonymous>'], inplace=True, errors='ignore')
# Â∞ÜÊâÄÊúâÁöÑ 'n/a' ÊõøÊç¢‰∏∫ÁúüÊ≠£ÁöÑÁ©∫ (NaN)
df_spotify.replace('n/a', np.nan, inplace=True)

# ÂÅáËÆæ‰Ω†ÊÉ≥ÊääÂêàÂπ∂ÂêéÁöÑÁªìÊûúÂ≠òÂÖ•Êñ∞Âàó 'final_city'
df_spotify['final_city'] = df_spotify['city_1'].combine_first(df_spotify['city_2']).combine_first(df_spotify['city_3'])

df_spotify.drop(columns=['city_1', 'city_2', 'city_3'], inplace=True)

df_spotify.columns = ['art_name', 'gender', 'age', 'art_type', 'area', 'city']

# ËΩ¨Êç¢‰∏∫Á±ªÂà´Âûã (Category)
df_spotify['gender'] = df_spotify['gender'].astype('category')

# È™åËØÅÁªìÊûú
print(df_spotify['gender'].dtype)

df_spotify.info()
```

```python
# 1. ÊîæÂÖ•‰Ω†ÊãÜÂàÜÂ•ΩÁöÑÂèòÈáèÂàóË°®
dataframes = {
    'Global Table': df_global,
    'Query Table': df_query,
    'Spotify Table': df_spotify
}

for name, df in dataframes.items():
    # ËÆ°ÁÆóÁº∫Â§±ÂÄºÂíåÂç†ÊØî
    null_stats = pd.DataFrame({
        'Missing Count': df.isnull().sum(),
        'Percentage (%)': (df.isnull().sum() / len(df) * 100).round(2)
    })

    # ÊâìÂç∞ÁæéÂåñË°®Â§¥
    print("\n" + "=" * 50)
    print(colored(f"üìä Quality Audit: {name}", 'cyan', attrs=['bold']))
    print(f"Total Rows: {len(df):,} | Columns: {len(df.columns)}")
    print("-" * 50)

    # Âè™ÊòæÁ§∫ÊúâÁº∫Â§±ÁöÑÂàó
    missing_only = null_stats[null_stats['Missing Count'] > 0].sort_values('Percentage (%)', ascending=False)

    if not missing_only.empty:
        print(missing_only)
    else:
        print(colored("‚úÖ Perfect: No missing values found.", 'green'))

print("=" * 50)
```

```python
# Âæ™ÁéØÊü•Áúã dataframes Â≠óÂÖ∏ÈáåÁöÑÊØè‰∏Ä‰∏™Ë°®
for name, df in dataframes.items():
    # 1. ÂàõÂª∫‰∏Ä‰∏™ÈÜíÁõÆÁöÑ HTML ËìùËâ≤Ê†áÈ¢ò
    display(HTML(f"""
        <div style="background-color: #e6f3ff; padding: 10px; border-radius: 5px; border-left: 5px solid #2196F3; margin-top: 20px;">
            <h3 style="color: #0d47a1; margin: 0;">üìä Dataset: {name}</h3>
            <p style="margin: 5px 0 0 0; color: #555;">Shape: {df.shape[0]:,} rows √ó {df.shape[1]} columns</p>
        </div>
    """))

    # 2. ÊâìÂç∞ Basic Info (Êï∞ÊçÆÁ±ªÂûã„ÄÅÈùûÁ©∫ÂÄºÁªüËÆ°)
    print(colored(f"\n>>> [{name}] Info:", 'blue', attrs=['bold']))
    df.info()

    # 3. ÁæéÂåñÂ±ïÁ§∫ Describe (Êï∞ÂÄºÂàóÁªüËÆ°ÊëòË¶Å)
    # ‰ΩøÁî® .T (ËΩ¨ÁΩÆ) ÊòØÂõ†‰∏∫Â¶ÇÊûúÂàóÊï∞ÂæàÂ§öÔºåÊ®™ÂêëÊéíÂàó‰ºöÂæàÈöæÁúãÔºåËΩ¨ÁΩÆÂêéÊõ¥Á¨¶ÂêàÈòÖËØª‰π†ÊÉØ
    print(colored(f"\n>>> [{name}] Statistical Summary (Transposed):", 'blue', attrs=['bold']))
    display(df.describe(include='all').T)

    print("-" * 80)
```

```python
# 1. È¢ÑÂ§ÑÁêÜÔºöÁîüÊàêÊ∏ÖÊ¥óÂêéÁöÑÂåπÈÖçÈîÆ
def create_artist_key(df, col_name):
    return df[col_name].astype(str).str.lower().str.replace(r'[^a-z0-9]', '', regex=True)


df_global['match_key'] = create_artist_key(df_global, 'art_name')
df_query['match_key'] = create_artist_key(df_query, 'art_name')
df_spotify['match_key'] = create_artist_key(df_spotify, 'art_name')

# 2. ËÆæÁΩÆÁ¥¢Âºï‰ª•‰æõ combine_first ÂØπÈΩê
df_global.set_index('match_key', inplace=True)
df_query.set_index('match_key', inplace=True)
df_spotify.set_index('match_key', inplace=True)

# 3. Ë°•ÈΩêÂêàÂπ∂Ôºö‰ª• Spotify ‰∏∫‰∏ªÔºå‰æùÊ¨°Âêë Query Âíå Global Ë¶ÅÊï∞ÊçÆ
df_final = df_spotify.combine_first(df_query).combine_first(df_global)

# 4. ÊÅ¢Â§çÁ¥¢ÂºïÂπ∂Ê∏ÖÁêÜ
df_final.reset_index(inplace=True)
```

```python
# 1. Áªü‰∏Ä‰∏ªÈîÆÂπ∂ÁîüÊàêÊ∏ÖÊ¥óÂêéÁöÑ match_key (ÂÖ≥ÈîÆÔºÅ)
def get_clean_key(df, col):
    return df[col].astype(str).str.lower().str.replace(r'[^a-z0-9]', '', regex=True)


df_global['m_key'] = get_clean_key(df_global, 'art_name')
df_query['m_key'] = get_clean_key(df_query, 'art_name')
df_spotify['m_key'] = get_clean_key(df_spotify, 'art_name')

# 1. Á°Æ‰øù‰∏âÂº†Ë°®ÈÉΩÊúâÁªü‰∏ÄÁöÑÊ∏ÖÊ¥óÈîÆ
# 2. „ÄêÂÖ≥ÈîÆÊ≠•È™§„ÄëÂ∞Ü m_key ËÆæ‰∏∫Á¥¢Âºï
df_g_idx = df_global.set_index('m_key')
df_q_idx = df_query.set_index('m_key')
df_s_idx = df_spotify.set_index('m_key')

# 3. Ê£ÄÊü•Á¥¢ÂºïÊòØÂê¶ÂîØ‰∏ÄÔºàÂ¶ÇÊûúËøîÂõû FalseÔºåËØ¥ÊòéÂçïË°®ÂÜÖÂ∞±ÊúâÈáçÂ§çÊ≠åÊâãÔºâ
print(f"Global Á¥¢ÂºïÂîØ‰∏ÄÂêó: {df_g_idx.index.is_unique}")

# 4. ÂÜçÊ¨°ÊâßË°åÂêàÂπ∂
# combine_first ‰ºöÊ†πÊçÆÂàöÊâçËÆæÂ•ΩÁöÑ index Ëá™Âä®Êâæ‰∫∫ÂØπÈΩê
df_final = df_s_idx.combine_first(df_q_idx).combine_first(df_g_idx)

# 5. Ê≠§Êó∂ÁúãË°åÊï∞ÔºåÂ∫îËØ•ËøúÂ∞è‰∫é 31‰∏á
df_final.reset_index(inplace=True)
print(f"ÁúüÊ≠£ÂêàÂπ∂ÂêéÁöÑÊ≠åÊâãÊÄªÊï∞: {len(df_final)}")
df_final.info()
```
